```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("dataset.csv",encoding='latin-1')
```

```python
date_cols = ['approxdate', 'resolution']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
missing_values=df.isnull().sum()
```

```python
# Distribution of Terrorist Attacks across Regions and Countries
plt.figure(figsize=(12, 6))
sns.countplot(x='region_txt', data=df, order=df['region_txt'].value_counts().index)
plt.title('Distribution of Terrorist Attacks across Regions')
plt.xlabel('Region')
plt.ylabel('No of Attacks')
plt.xticks(rotation=50)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_countries = df['country_txt'].value_counts().nlargest(15)
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title('Top 10 Most Targeted Countries')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=50)
plt.show()
```

```python
# Trend of Terrorist Attacks Over the Years
plt.figure(figsize=(10, 5))
df['year'] = df['iyear'] + df['imonth'] / 12
sns.lineplot(x='year', y='iyear', data=df)
plt.title('Trend of Terrorist Attacks Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Most Common Attack Types
plt.figure(figsize=(12, 6))
sns.countplot(x='attacktype1_txt', data=df, order=df['attacktype1_txt'].value_counts().index)
plt.title('Most Common Attack Types')
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.xticks(rotation=50)
plt.show()
```

```python
# Casualties and Injuries Analysis
plt.figure(figsize=(12, 6))
sns.scatterplot(x='nkill', y='nwound', data=df, hue='attacktype1_txt', palette='viridis', size='nkill')
plt.title('Casualties and Injuries Analysis')
plt.xlabel('No of Killed')
plt.ylabel('No of Wounded')
plt.legend(title='Attack Type', bbox_to_anchor=(1,1))
plt.show()
```

```python
# Analysis of Specific Terrorist Groups
plt.figure(figsize=(12, 6))
top_terrorist_groups = df['gname'].value_counts().nlargest(10)
sns.barplot(x=top_terrorist_groups.index, y=top_terrorist_groups.values, palette='Set3')
plt.title('Top 10 Most Active Terrorist Groups')
plt.xlabel('Terrorist Groups')
plt.ylabel('No of Attacks')
plt.xticks(rotation=50)
plt.show()
```

```python
# Correlation Analysis
correlation_matrix = df.corr()

# Display Correlation Matrix as a Table
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
# Statistical Analysis
# Example: Hypothesis Testing
from scipy.stats import ttest_ind
# Hypothesis: Is there a significant difference in the number of killed between successful and unsuccessful attacks?
successful_attacks = df[df['success'] == 1]['nkill']
unsuccessful_attacks = df[df['success'] == 0]['nkill']
t_statistic, p_value = ttest_ind(successful_attacks.dropna(), unsuccessful_attacks.dropna(), equal_var=False)
print(f"T-Statistic: {t_statistic}, p-value: {p_value}")
```
